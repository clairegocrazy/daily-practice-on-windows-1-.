```python
#撰寫一個大樂透電腦自動選號程式,程式的執行會以亂數顯示1-49之間7個不重複的大樂透號碼
import random
biglottery=set()
while (len(biglottery)<7):
    biglottery.add(random.randint(1,49))
lottery_list=list(biglottery)
print("本組號碼:",end='')
print(sorted(lottery_list[:6]),end='')
print("特別號:",end='')
print(lottery_list[-1])
```

```python
import random
def generate_lottery():
    biglottery=set()
    while (len(biglottery)<7):
         biglottery.add(random.randint(1,49))
    lottery_list=list(biglottery)
    print("本組號碼:",end='')
    print(sorted(lottery_list[:6]),end='')
    print("特別號:",end='')
    print(lottery_list[-1])
nums=int(input("請輸入大樂透電腦選號的組數:"))
for i in range (nums):
    print(f"第{i+1}組")
    generate_lottery()
```
